```python
# 1. 군집
# 비슷한 샘플을 구별해 하나의 클러스터 또는 비슷한 샘플의 그룹으로 할당하는 작업
# 종류 - 고객분류, 데이터분석, 차원축소기법, 이상치탐지, 준지도학습, 검색엔진, 이미지분할

# 1.1 k-평균
# 데이터 생성
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

# 군집의 중심점 정의
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])

# 군집에서 데이터간 정규화 값 정의(간격)
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

# 군집당 샘플수, 중심값, 정규화 값, 랜덤값 정의
X, y = make_blobs(n_samples=2000, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)


def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1) # s는 점의 굵기, c는 라벨로 생각
    plt.xlabel("$x_1$", fontsize=14)
    plt.ylabel("$x_2$", fontsize=14, rotation=0)
    
plt.figure(figsize=(12, 8))
plot_clusters(X)
plt.show()
```

```python
# 훈련
# 클러스터의 중심을 찾고 가장 가까운 클러스터에 샘플 할당

from sklearn.cluster import KMeans
# 클러스터(무리)수 정의
# 위의 그래프에서 군집 수는 5
k=5
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(X)
y_pred
```

```python
# k-평균 알고리즘의 원리
# 1. 랜덤하게 센트로이드 초기화
# 2. 샘플에 레이블 할당
# 3. 센트로이드 업데이트
# 4. 샘플에 다시 레이블 할당
# 5. 센트로이드에 변화가 없을때까지 반복
```

```python
# 센트로이드 초기화 방법
# 1. 샌트로이드 위치를 근사하게 알 수 있는 경우 근사값으로 초기화

good_init = np.array([[-3, 3], [-3, 2], [-3, 1], [-1, 2], [0, 2]])
kmeans = KMeans(n_clusters=5, init=good_init, n_init=1)
```

```python
# 2. 랜덤초기화를 다르게 하여 여러번 알고리즘을 실행
# 랜덤초기화 횟수는 n_init 값을 조정, default = 10
# 기본값 10번의 랜덤 초기화를 해서 최선의 솔류션을 선택
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
```

```python
# 이니셔
# 최선의 솔루션을 선택하는 방법
# 값이 작을수록 좋은 솔류션
kmeans.inertia_
```

```python
# 군집을 찾아가는 과정

kmeans_iter1 = KMeans(n_clusters=5, init="random", n_init=1,
                      algorithm="full", max_iter=1, random_state=0)
kmeans_iter2 = KMeans(n_clusters=5, init="random", n_init=1,
                      algorithm="full", max_iter=2, random_state=0)
kmeans_iter3 = KMeans(n_clusters=5, init="random", n_init=1,
                      algorithm="full", max_iter=3, random_state=0)
kmeans_iter4 = KMeans(n_clusters=5, init="random", n_init=1,
                      algorithm="full", max_iter=4, random_state=0)
kmeans_iter1.fit(X)
kmeans_iter2.fit(X)
kmeans_iter3.fit(X)
kmeans_iter4.fit(X)

def draw_kmeans(kmeans):
    x0, x1 = np.meshgrid(
        np.linspace(-3.5, 1.5, 400).reshape(-1, 1),
        np.linspace(0.5, 4, 300).reshape(-1, 1)
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = kmeans.predict(X_new)
    zz = y_predict.reshape(x0.shape)
    plt.contourf(x0, x1, zz, cmap="Pastel2")
    plt.scatter(X[:, 0], X[:, 1], s=1, color="black")

    y_pred = kmeans.predict(X_new).reshape(x1.shape)
    plt.contour(x0, x1, y_pred, colors="k", alpha=0.8)

    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                    marker='o', s=35, linewidths=8,
                    color="w", zorder=10, alpha=0.9)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                    marker='x', s=2, linewidths=12, 
                    color="k", zorder=11, alpha=1)

plt.figure(figsize=(12, 8))
plt.subplot(221);
draw_kmeans(kmeans_iter1)
plt.title("1", fontsize=12)
plt.subplot(222);
draw_kmeans(kmeans_iter2)
plt.title("2", fontsize=12)
plt.subplot(223);
draw_kmeans(kmeans_iter3)
plt.title("3", fontsize=12)
plt.subplot(224);
draw_kmeans(kmeans_iter4)
plt.title("4", fontsize=12)
```

```python
# 이니셔 비교
for i in range (1, 10):
    kmeans_iter = KMeans(n_clusters=5, init="random", n_init=1,
                      algorithm="full", max_iter=i, random_state=0)
    kmeans_iter.fit(X)
    print("Sequence : {}".format(i))
    print("Inertia : {}".format(kmeans_iter.inertia_))
```

```python
# 미니배치 K-평균
# 각 반복마다 미니배치를 사용해 센트로이드를 조금씩 이동한다.

from sklearn.cluster import MiniBatchKMeans

minibatch_kmeans = MiniBatchKMeans(n_clusters = 5)
minibatch_kmeans.fit(X)
```

```python
# 미니배치 K-평균과 일반 K-평균 비교
# 1. 이니셔
# 일반적으로 일반 K-평균이 점수가 높음

mini_inertia = []
nomal_inertia = []

for i in range(1, 100):
    minibatch_kmeans = MiniBatchKMeans(n_clusters = i)
    minibatch_kmeans.fit(X)
    mini_inertia.append(minibatch_kmeans.inertia_)
    kmeans_iter = KMeans(n_clusters=i, init="random", n_init=1,
                      algorithm="full", max_iter=i, random_state=0)
    kmeans_iter.fit(X)
    nomal_inertia.append(kmeans_iter.inertia_)


X_axis = np.linspace(1, 100, 99)
plt.figure(figsize=(12, 8))
plt.plot(X_axis, mini_inertia, "b-", label="Mini K-average")
plt.plot(X_axis, nomal_inertia, "r-", label="K-average")
plt.axis([0, 100, 0, 100])
plt.legend(fontsize=13)
plt.xlabel("Number of cluster(centroid)", fontsize=13)
plt.ylabel("Score of Inertia", fontsize=13)
```

```python
# 2. 훈련시간
# 미니배치가 더 빠름

from timeit import timeit
times = np.empty((100, 2))
inertias = np.empty((100, 2))
for k in range(1, 101):
    kmeans_ = KMeans(n_clusters=k, random_state=42)
    minibatch_kmeans = MiniBatchKMeans(n_clusters=k, random_state=42)
    print("\r{}/{}".format(k, 100), end="")
    times[k-1, 0] = timeit("kmeans_.fit(X)", number=10, globals=globals())
    times[k-1, 1]  = timeit("minibatch_kmeans.fit(X)", number=10, globals=globals())
    inertias[k-1, 0] = kmeans_.inertia_
    inertias[k-1, 1] = minibatch_kmeans.inertia_
    
plt.plot(range(1, 101), times[:, 0], "r--", label="K-Means")
plt.plot(range(1, 101), times[:, 1], "b.-", label="Mini-batch K-Means")
plt.xlabel("$k$", fontsize=16)
plt.title("Training time (seconds)", fontsize=14)
plt.axis([1, 100, 0, 6])
```

```python
# 최적의 클러스터(센트로이드)개수 설정
# 1. 이니셔를 기준으로? 이니셔는 클러스터의 수가 늘어남에 따라 줄어듦 따라서 좋은 지표가 아님
# 하지만 갑자기 꺾이는 부분(Elbow)를 기준으로 하는것은 유용

X_axis = np.linspace(1, 9, 9)
print(X_axis)
y_axis = []
for i in range(1, 10):
    y_axis.append(mini_inertia[i])

plt.figure(figsize=(12, 8))
plt.plot(X_axis, y_axis, "r-")
plt.axis([0, 10, 0, 1200])
plt.xlabel("Number of cluster(centroid)", fontsize=13)
plt.ylabel("Score of Inertia", fontsize=13)
plt.annotate('Elbow',
             xy=(3, y_axis[2]),
             xytext=(0.55, 0.55),
             textcoords='figure fraction',
             fontsize=16,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
```

```python
# 2. 실루엣 점수 활용
# 1에 가까울수록 자신의 클러스터 안에 잘 속해 있고 다른 클러스터와는 멀리 떨어져 있음
# 0에 가까울수록 클러스터의 경계에 위치
# -1에 가까울수록 샘플이 잘못된 클러스터에 할당되었다는 의미

from sklearn.metrics import silhouette_score
silhouette_score(X, kmeans.labels_)
```

```python
# 클러스터에 따른 실루엣 점수
# 엘보우는 3이지만 실루엣 점수는 4와 5가 좋아보임

score = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    print("\r{}/{}".format(k, 10), end="")
    score.append(silhouette_score(X, kmeans.labels_))

X_axis = np.linspace(2, 9, 8)
plt.figure(figsize = (16, 8))
plt.plot(X_axis, score)
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
import matplotlib as mpl

plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)
    
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    y_pred = kmeans.labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")
    
    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=score[k - 2], color="red", linestyle="--") # 클러스터 k에 해당하는 실루엣 점수
    plt.title("$k={}$".format(k), fontsize=16)

plt.show()

# 그래프 해석
# 1. 가로길이 -> 실루엣 계수
# 2. 세로길이 -> 클러스터에 포함된 샘플의 수
# 3. 빨간점선 -> 해당 클러스터 수에 따른 평균 실루엣 점수
# 빨간점선에 못미칠수록 클러스터의 샘플이 다른 클러스터랑 너무 가까움을 의미
# 클러스터가 모두 빨간점선을 넘을수록, 클러스터의 크기가 잘 분배될수록 좋은 클러스터의 수로 판단
```

```python
# 1.2 군집을 사용한 이미지 분할
# 세그먼트 -> 모음이라고 생각, 예를들어, 사람, 자동차, 꽃
# 색상분할 -> 동일한 색상을 가진 픽셀을 같은 세그먼트에 할당


# 데이터 준비
import urllib
images_path = os.path.join("images", "unsupervised_learning") # 파일경로
os.makedirs(images_path, exist_ok=True) # 없을시 생성
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/" # 다운로드 루트
filename = "ladybug.png" # 다운로드 루트에 위치한 파일 이름
print("Downloading", filename)
url = DOWNLOAD_ROOT + "images/unsupervised_learning/" + filename # 다운로드 루투 풀내임
urllib.request.urlretrieve(url, os.path.join(images_path, filename)) 
# 지정한 url에서 파일을 다운받아 지정한 경로에 지정한 이름으로 저장


# 데이터 읽기
import os
from matplotlib.image import imread
image = imread(os.path.join("images", "unsupervised_learning", "ladybug.png"))
image.shape
# 높이, 너비, 컬러채널개수(RGB : 0-1), from imageio import imread 사용시 0-255
# 흑백일시 더 적은 채널, 투명도를 가질시 더 많은 채널(알파채널)
```

```python
# 배열을 RGB 색상의 긴 리스트로 변환한 후 k-평균을 사용해 이 색상을 클러스터로 모음
# 예를들어, 밝은 노란색, 어두운 노란색 등을 평균적인 노란색으로 변환
# 여기서 클러스터는 색상의 수

X = image.reshape(-1, 3)
kmeans = KMeans(n_clusters=8).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
# 1.3 군집을 사용한 전처리
# 0-9까지의 숫자를 나타내는 8*8흑백이미지를 군집전처리를 통해 정확도를 높임

from sklearn.datasets import load_digits

X_digits, y_digits = load_digits(return_X_y=True)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
# 전처리 단계에 군집 추가시 점수 올라감을 확인
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=50)), # 8*8=64에서 50으로 감소
    ("log_reg", LogisticRegression()),
])
pipeline.fit(X_train, y_train)
pipeline.score(X_test, y_test)
```

```python
# 최적의 k 찾기

from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=range(2, 100)) # 클러스터를 2부터 99까지
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2) # 3분할하여 그리드서치 함
grid_clf.fit(X_train, y_train) # 훈련
```

```python
grid_clf.best_params_ # 모델에 저장된 최적의 파라미터(k) : 89
```

```python
grid_clf.score(X_test, y_test) # 점수가 더 높아짐
```

```python
# 1.4 군집을 사용한 준지도 학습
# 준지도 -> 레이블이 조금만 있고 대부분의 데이터가 레이블이 없을때

# 우선, 레이블이 된 50개의 데이터에 대한 훈련후 점수 평가
n_labeled = 50
log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
log_reg.score(X_test, y_test)
```

```python
# 과정
# 1. 훈련세트를 50개의 세그먼트(클러스터)로 지정
# 2. 각 클러스터에서 센트로이드에 가장 가까운 이미지를 찾음 -> 대표이미지
# 출력 : 50개의 대표이미지

k=50
kmeans = KMeans(n_clusters=k)
X_digits_dist = kmeans.fit_transform(X_train) # 50개의 클러스터로 훈련데이터 셋을 나눔
representative_digit_idx = np.argmin(X_digits_dist, axis=0) # 대표이미지의 인덱스
X_representative_digits = X_train[representative_digit_idx] # 대표이미지 모음

plt.figure(figsize=(10, 8))
for index, X_representative_digit in enumerate(X_representative_digits):
    # enumerate -> 지정한 변수의 값뿐만 아니라 인덱스도 반환
    plt.subplot(k // 10, 10, index + 1) # 10개씩 10줄
    plt.imshow(X_representative_digit.reshape(8, 8), cmap="binary", interpolation="bilinear")
    plt.axis('off')

plt.show()
```

```python
# 레이블 전파
# 위에서 생성한 대표이미지의 레이블을 나머지 데이터에도 전파하는 과정

y_train_propagated = np.empty(len(X_train), dtype=np.int32) 
# 훈련세트의 크기만큼 int32에 해당하는 숫자를 채운 배열 만듦
# 1347개의 데이터
for i in range(k): # k는 50
    y_train_propagated[kmeans.labels_==i] = y_representative_digits[i]
    # 레이블 전파
    
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

```python
# 점수가 그렇게 좋지 않은 이유
# 경계에 걸친 데이터 때문에
# 센트로이드와 가까운 샘풀 20%만 레이블을 전파로 해결해 보자

percentile_closet = 20
X_cluster_dist = X_digits_dist[np.arange(len(X_train)), kmeans.labels_]
for i in range(k):
    in_cluster = (kmeans.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closet)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1
    
partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
X2 = X2 + [6, -8]
X = np.r_[X1, X2]
y = np.r_[y1, y2]

# 기본 데이터 표현
plt.figure(figsize=(12, 8))
plt.scatter(X[:, 0], X[:, 1], marker="x")

x0s = np.linspace(-4, 7, 100)
x1s = np.linspace(-3, 5.5, 100)
x0, x1 = np.meshgrid(x0s, x1s)
X = np.c_[x0.ravel(), x1.ravel()]


y_pred = gm.predict(X).reshape(x0.shape) # 예측
y_decision = gm.score_samples(X).reshape(x0.shape) # 결정함수

# 예측에 대한 경계선 
plt.contour(x0, x1, y_pred, colors="r", linewidth=2, linestyles="dashed", alpha=0.8)

# 결정함수에 따른 등고선
plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.2)


# 클러스터의 중심(평균)
plt.scatter(gm.means_[:, 0], gm.means_[:, 1],
                marker='o', s=35, linewidths=8,
                color="w", zorder=10, alpha=0.9)
plt.scatter(gm.means_[:, 0], gm.means_[:, 1],
                marker='x', s=2, linewidths=12, 
                color="k", zorder=11, alpha=1)
```

```python
# 클러스터의 모양이나 범위 제한
# 특성이나 클러스터가 많거나 샘플이 적을때 사용

gm_full = GaussianMixture(n_components=3, n_init=10, covariance_type="full", random_state=42)
gm_tied = GaussianMixture(n_components=3, n_init=10, covariance_type="tied", random_state=42)
gm_spherical = GaussianMixture(n_components=3, n_init=10, covariance_type="spherical", random_state=42)
gm_diag = GaussianMixture(n_components=3, n_init=10, covariance_type="diag", random_state=42)
gm_full.fit(X)
gm_tied.fit(X)
gm_spherical.fit(X)
gm_diag.fit(X)


def plot_something(gm):
    
    X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
    X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
    X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
    X2 = X2 + [6, -8]
    X = np.r_[X1, X2]
    y = np.r_[y1, y2]

    # 기본 데이터 표현
    plt.scatter(X[:, 0], X[:, 1], marker="x")

    x0s = np.linspace(-4, 7, 100)
    x1s = np.linspace(-3, 5.5, 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]


    y_pred = gm.predict(X).reshape(x0.shape) # 예측
    y_decision = gm.score_samples(X).reshape(x0.shape) # 결정함수

    # 예측에 대한 경계선 
    plt.contour(x0, x1, y_pred, colors="r", linewidth=2, linestyles="dashed", alpha=0.8)

    # 결정함수에 따른 등고선
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.2)


    # 클러스터의 중심(평균)
    plt.scatter(gm.means_[:, 0], gm.means_[:, 1],
                    marker='o', s=35, linewidths=8,
                    color="w", zorder=10, alpha=0.9)
    plt.scatter(gm.means_[:, 0], gm.means_[:, 1],
                    marker='x', s=2, linewidths=12, 
                    color="k", zorder=11, alpha=1)
    

plt.figure(figsize=(12, 10))
plt.subplot(221);
plot_something(gm_full)
plt.subplot(222);
plot_something(gm_tied)
plt.subplot(223);
plot_something(gm_spherical)
plt.subplot(224);
plot_something(gm_diag)
```

```python
# 가우시안 혼합의 이상치 탐지
# 밀도가 낮은 지역에 있는 모든 샘플을 이상치로 탐지
# 기준 임계값 지정 후 그것보다 낮으면 이상치르 판단
X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
X2 = X2 + [6, -8]
X = np.r_[X1, X2]
y = np.r_[y1, y2]
densities = gm.score_samples(X)
density_threshold = np.percentile(densities, 4) # 하위 4%를 임계값으로 지정
anomalies = X[densities < density_threshold] # 임계값보다 낮은 데이터
plt.figure(figsize=(12, 8))
plot_something(gm)
plt.scatter(anomalies[:, 0], anomalies[:, 1], marker="x", c="r", s=100)
```
