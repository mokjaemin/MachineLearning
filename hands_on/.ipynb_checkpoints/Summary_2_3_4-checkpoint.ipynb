```python
# 변환기 만들기

from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6
# 실제 데이터의 인덱스임. 인덱스 3은 방수, 4는 화장실수..

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    # BaseEstimator 상속시 하이퍼 파라미터 튜닝에 필요한 메서드 get_params, set_params 자동 추가
    # TransformerMixin  상속시 자동으로 fit_transform 메서드 자동 생성
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
        # 이 변환기의 하이퍼 파라미터는 add_bedrooms_per_room 하나.
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix] # 방수를 가구수로 나눔, 가구 당 방수
        population_per_household = X[:, population_ix] / X[:, households_ix] # 인구수를 가구수로 나눔, 가구당 인구수
        
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix] # 방당 화장실 수
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
            # 방당 화장실 수를 추가해서 반환
        else:
            return np.c_[X, rooms_per_household, population_per_household]
            # 방당 화장실 수를 제외하고 반환
        
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
# 숫자 그림을 위한 추가 함수

def plot_digits(instances, images_per_row=10, **options): # 데이터, 줄당 이미지 수
    size = 28
    
    # 100과 10사이의 최솟값
    images_per_row = min(len(instances), images_per_row) 

    
    # n_rows = ceil(len(instances) / images_per_row) 와 동일합니다 
    # 즉, 한줄당 이미지가 10개들어가므로 10줄을 의미
    n_rows = (len(instances) - 1) // images_per_row + 1 # 10

    
    # 필요하면 그리드 끝을 채우기 위해 빈 이미지를 추가합니다:
    # 예를 들어, 데이터가 99개이고 줄당 데이터가 10개, 10줄 존재시 n_empty = 1
    n_empty = n_rows * images_per_row - len(instances)
    
    
    # concatenate -> 배열을 합침, 기존의 데이터에 0으로가득찬 데이터를 이미지크기만큼, 빈 사진만큼 가로열에 추가 
    # padded_instances -> (100, 784)
    padded_instances = np.concatenate([instances, np.zeros((n_empty, size * size))], axis=0)

    
    # 배열의 크기를 바꾸어 28×28 이미지를 담은 그리드로 구성합니다:
    # 줄 수, 줄 당 이미지 수, 사진의 가로, 세로
    
    # 형태
    # 이미지1 ~ 이미지10
    # 이미지11 ~ 이미지 20
    # ...
    image_grid = padded_instances.reshape((n_rows, images_per_row, size, size))

    
    
    # 축 0(이미지 그리드의 수직축)과 2(이미지의 수직축)를 합치고 축 1과 3(두 수평축)을 합칩니다. 
    # 먼저 transpose()를 사용해 결합하려는 축을 옆으로 이동한 다음 합칩니다:
    # reshape(280, 280)을 통해 하나의 사진으로 만듦
    big_image = image_grid.transpose(0, 2, 1, 3).reshape(n_rows * size, images_per_row * size)
    
    
    # 하나의 큰 이미지를 얻었으므로 출력하면 됩니다:
    plt.imshow(big_image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
```

```python
# 훈련, 테스트 데이터 분리

X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
# 이진 분류기

y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)

from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier()
sgd_clf.fit(X_train, y_train_5)
```

```python
sgd_clf.predict([X[0]]) # 2차원, 일렬로 된 데이터를 입력으로 줄것
```

```python
# 성능 지표

# 1. 교차 검증 1
# cross_val_score 구현해보기

from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone


sgd_clf = SGDClassifier()
skfolds = StratifiedKFold(n_splits = 3, random_state = 42, shuffle = True)
# 3개의 폴드로 나누고 계층적 샘플링 수행함


for train_index, test_index in skfolds.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)
    
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_test_fold = X_train[test_index]
    y_test_fold = y_train_5[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    test_predict = clone_clf.predict(X_test_fold)
    correct = sum(test_predict == y_test_fold)
    print("cross val score : ", correct/len(test_predict))
```

```python
# 1. 교차 검증 2
# 이미 구현된 교차검증 cross_val_score 사용

from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

```python
# 2. 오차행렬
# 정밀도와 재현율, F1점수로 성능 판단

from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

# cross_val_predict -> 교차검증을 수행하지만 평가점수를 반환하지 않고 예측 결과를 반환
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
confusion_matrix(y_train_5, y_train_pred)
```

```python
# 정밀도
from sklearn.metrics import precision_score
precision_score(y_train_5, y_train_pred)
```

```python
# 재현율
from sklearn.metrics import recall_score
recall_score(y_train_5, y_train_pred)
```

```python
# F1 점수
from sklearn.metrics import f1_score
f1_score(y_train_5, y_train_pred)
```

```python
# 임계값에 따른 정밀도와 재현율 표현

from sklearn.metrics import precision_recall_curve
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="recall")
    # [:-1] -> precisions, recalls 가 데이터가 하나가 더 많음 따라서 마지막 데이터 한개 삭제
    
    a_threshold = 8000
    plt.plot(a_threshold, )
    plt.xlabel("threshold")
    plt.axis([-50000, 50000, 0, 1])
    plt.legend()
    plt.grid()

plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

```python
# 정밀도와 재현율 그래프

plt.plot(recalls, precisions, "b--")
plt.xlabel("recall")
plt.ylabel("precision")
plt.grid()
```

```python
# 정밀도가 90%이상될때 임계값

threshold_90_precision = thresholds[np.argmax(precisions >= 0.9)]
threshold_90_precision
```

```python
# 임계값에 해당하는 정밀도와 재현율 계산

y_train_pred_90 = (y_scores >= threshold_90_precision)
```

```python
# 정밀도
precision_score(y_train_5, y_train_pred_90)
```

```python
# 재현율

recall_score(y_train_5, y_train_pred_90)
```

```python
# 4장 정리

# 데이터 생성
import numpy as np
X = 2*np.random.rand(100, 1) # (100,1), 0~2
y = 4 + 3*X + np.random.randn(100, 1)
```

```python
# 경사 하강법
# 1. 배치 경사 하강법

def graph_per_eta(eta): # eta - 학습률
    
    n_iterations = 1000
    m = 100 # 샘플수

    theta = np.random.randn(2, 1) # 무작위 초기화
    theta_house = []

    for iteration in range(n_iterations):
        gradient = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * gradient
        theta_house.append(theta)
        y_predict = []

    # 5개의 스텝을 담음
    for i in range(0, 5):
        y_predict.append(X_new_b.dot(theta_house[i]))

    for i in range(0, 5):
        plt.plot(X_new, y_predict[i], "r-")

    plt.plot(X, y, "b.")
    plt.axis([0, 2, 0, 14])
```

```python
# 5개의 스텝
# 학습률 0.02, 0.1, 0.5 

plt.figure(figsize=(10,10))
plt.subplot(221);
graph_per_eta(0.02)
plt.subplot(222);
graph_per_eta(0.1)
plt.subplot(223);
graph_per_eta(0.5)
```

```python
# 2. 확률적 경사 하강법(SGD)

n_epochs = 50
m = 100
t0, t1 = 5, 50

def learning_schedule(t):
    return t0/(t+t1)

theta = np.random.randn(2, 1) # 무작위 초기화
theta_house_epoch = []

for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index : random_index + 1]
        yi = y[random_index : random_index + 1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_house_epoch.append(theta)
```

```python
# 에포크의 20개의 스텝
for i in range(0, 20):
    y_predict.append(X_new_b.dot(theta_house_epoch[i]))

for i in range(0, 20):
    plt.plot(X_new, y_predict[i], "r-")

plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 14])
```

```python
# 사이킷런으로 확률적 경사하강법 구현

from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1)
# 에포크는 1000, 한 에포크당 손실이 0.001보다 적게, 학습률은 0.1
sgd_reg.fit(X, y.ravel())
# ravel() -> 2차원을 1차원으로 변경
```

```python
# 전처리
# 훈련세트에 있는 각 특성을 제곱하여 새로운 특성으로 추가

from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False) # 2차 다항 
X_poly = poly_features.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# 예측 그래프를 그리기 위한 데이터 생성
X_new = np.linspace(-3, 3, 100)
y_new = lin_reg.coef_[0][1]*X_new**2 + lin_reg.coef_[0][0]*X_new + lin_reg.intercept_
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

X_new = np.linspace(-3, 3, 100).reshape(100, 1)

for style, width, degree in (("g-", 1, 300),("b--", 2, 2),("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree = degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
        ("poly_features", polybig_features),
        ("std_scaler", std_scaler),
        ("lin_reg", lin_reg),
    ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig, style, linewidth=width, label=str(degree))
    
plt.plot(X, y, "b.")
plt.axis([-3, 3, 0, 10])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend()
```

```python
# 학습 곡선
# 훈련세트의 크기에 대비한 RMSE 평가
# 1차(직선)다항

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend()
    plt.xlabel("size of Training set")
    plt.ylabel("RMSE", rotation=0)

lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
```

```python
# 10차 다항으로 학습곡선 그려보기

from sklearn.pipeline import Pipeline
polynomial_regression = Pipeline([
    ("poly_features", PolynomialFeatures(degree=10, include_bias = False)),
    ("lin_reg", LinearRegression()),
])
plt.axis([0, 80, 0, 3])
plot_learning_curves(polynomial_regression, X, y)
```

```python
# 규제
# 릿지 회귀
# 10차 다항에 규제한 후 그래포
from sklearn.linear_model import Ridge


m = 20
X = 3*np.random.rand(m, 1)
y = 1 + 0.5*X + np.random.randn(m, 1)

X_new = np.linspace(0, 3, 100).reshape(100, 1)

for alpha, style in ((0, "b-"), (10**-5, "g--"), (1, "r:")):
    std_caler = StandardScaler()
    polynomial_features = PolynomialFeatures(degree=10, include_bias=False)
    ridge_reg = Ridge(alpha=alpha, solver="cholesky")
    pipeline = Pipeline([
        ("std_scaler", std_caler),
        ("polynomial_features", polynomial_features),
        ("ridge_reg", ridge_reg),
    ])
    pipeline.fit(X, y)
    y_new = pipeline.predict(X_new)
    plt.plot(X_new, y_new, style, linewidth=2, label=str(alpha))
    
plt.plot(X, y, "b.")
plt.axis([0, 3, 0, 4])
# 규제가 커질슈록 완만해짐
```
