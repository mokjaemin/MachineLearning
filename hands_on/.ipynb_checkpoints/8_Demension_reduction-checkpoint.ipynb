```python
# 1. 투영 -> 그림자로 생각
# 2. 매니폴드 학습 -> 펼친다고 생각
# 3. PCA(주성분 분석)
# 데이터에 가장 가까운 초평면을 정의한 다음, 데이터를 이 평면에 투영
# 분산보존 -> 저차원의 초평면에 훈련세트를 투영시킬때 최대한 분산(데이터를 최대한 많이 표현)을 유지함.
# 주성분 -> 첫번째 주성분(PC)는 분산을 최대한 보존하는 벡터, 두번째는 이와 직교하고 남은 분산 최대한 보존하는 벡터, 이후 순차적 진행
# d차원으로 투영 -> 생선한 주성분을 토대로 d차원으로 투영(ex> 3차원->2차원)
```

```python
# 3차원 데이터 셋 생성

np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 3))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 0] * w1 + X[:, 1] * w2 + noise * np.random.randn(m)
```

```python
# 주성분(PC) 찾기
# 처음 두개의 PC를 추출 c1,c2

X_centered = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]
```

```python
# d차원으로 투영하기

W2 = Vt.T[:, :2] # 위의 c1,c2 (PC)
X2D = X_centered.dot(W2)
```

```python
# 사이킬런을 사용하여 주성분 분석을 통해 데이터셋의 차원을 3차원에서 2차원으로 줄임

from sklearn.decomposition import PCA

pca = PCA(n_components=2) # 2차원
X2D = pca.fit_transform(X)

pca.components_.T[:, 0] # 첫번째 PC, 벡터
```

```python
# 설명된 분산의 비율
# 각 PC에 담긴 분산의 비율

pca.explained_variance_ratio_
# 첫번째 PC에 84퍼, 두번쨰 PC에 14퍼, 대략 2퍼정도 세번째 PC
```

```python
# 지금부터의 그래프들은 참조만 하자. 이해하는데 도움이 되게만!

# 3D 화살표 그리기

from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, self.axes.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

```python
# X와 y의 함수로 평면을 표현

axes = [-1.8, 1.8, -1.3, 1.3, -1.0, 1.0]

x1s = np.linspace(axes[0], axes[1], 10) # 축 X1 => -1.8 ~ 1.8, 10개로 표현
x2s = np.linspace(axes[2], axes[3], 10) # 축 X2 => -1.3 ~ 1.3, 10개로 표현
x1, x2 = np.meshgrid(x1s, x2s) # 축 X1과 X2를 같이 표현

C = pca.components_ # 벡터(PC)
R = C.T.dot(C)
z = (R[0, 2] * x1 + R[1, 2] * x2) / (1 - R[2, 2])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

X3D_inv = pca.inverse_transform(X2D) # 위에서 생성한 2D를 다시 3D로 복원

fig = plt.figure(figsize=(12, 8)) # 그래프 사이즈 설정
ax = fig.add_subplot(111, projection='3d') # subplot 변수 생성

X3D_above = X[X[:, 2] > X3D_inv[:, 2]]
X3D_below = X[X[:, 2] <= X3D_inv[:, 2]]

ax.plot(X3D_below[:, 0], X3D_below[:, 1], X3D_below[:, 2], "bo", alpha=0.5)

ax.plot_surface(x1, x2, z, alpha=0.2, color="k")
np.linalg.norm(C, axis=0)
ax.add_artist(Arrow3D([0, C[0, 0]],[0, C[0, 1]],[0, C[0, 2]], mutation_scale=15, lw=1, arrowstyle="-|>", color="k"))
ax.add_artist(Arrow3D([0, C[1, 0]],[0, C[1, 1]],[0, C[1, 2]], mutation_scale=15, lw=1, arrowstyle="-|>", color="k"))
ax.plot([0], [0], [0], "k.")

for i in range(m):
    if X[i, 2] > X3D_inv[i, 2]:
        ax.plot([X[i][0], X3D_inv[i][0]], [X[i][1], X3D_inv[i][1]], [X[i][2], X3D_inv[i][2]], "k-")
        pass
    else:
        ax.plot([X[i][0], X3D_inv[i][0]], [X[i][1], X3D_inv[i][1]], [X[i][2], X3D_inv[i][2]], "-", color="#505050")
    
ax.plot(X3D_inv[:, 0], X3D_inv[:, 1], X3D_inv[:, 2], "k+")
ax.plot(X3D_inv[:, 0], X3D_inv[:, 1], X3D_inv[:, 2], "k.")
ax.plot(X3D_above[:, 0], X3D_above[:, 1], X3D_above[:, 2], "bo")
ax.set_xlabel("$x_1$", fontsize=18, labelpad=10)
ax.set_ylabel("$x_2$", fontsize=18, labelpad=10)
ax.set_zlabel("$x_3$", fontsize=18, labelpad=10)
ax.set_xlim(axes[0:2])
ax.set_ylim(axes[2:4])
ax.set_zlim(axes[4:6])

plt.show()
```

```python
# 적절한 차원 수 선택하기
# 차원을 줄일때는 일반적으로 2, 3차원으로 줄이는게 일반적
# 하지만 분산을 고려해서 차원을 설정해보자!는 취지

from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, as_frame=False)
mnist.target = mnist.target.astype(np.uint8)

from sklearn.model_selection import train_test_split

X = mnist["data"]
y = mnist["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y)

pca = PCA() # 차원의 수를 임의로 지정하지 않음
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_) # 주성분이 늘어남에 따라 분산값을 순차적으로 저장
d = np.argmax(cumsum >= 0.95) + 1 # 분산이 0.95이상일 때의 차원의 수

d # 154차원일때 분산이 0.95이상으로 유지된다.
```

```python
plt.figure(figsize=(12, 8))
plt.plot(cumsum)
plt.grid()
plt.xlabel("Number of Demension", fontsize=13)
plt.ylabel("Explained Variance Ratio", fontsize=13)
plt.plot([0, d], [0.95, 0.95], "k:")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot(d, 0.95, "ko")
plt.axis([0, 400, 0, 1])

# 그래프 분석
# 차원을 약 100으로 축소해도 분산의 큰 손해가 없다.
```

```python
# 데이터 압축과 복원
# 복원시 재구성 오차(복원데이터와 기존데이터의 차이)가 존재

pca = PCA(n_components=154)
X_reduced = pca.fit_transform(X_train) # 차원 축소(압축)
X_recovered = pca.inverse_transform(X_reduced) # 차원 복원
```

```python
import matplotlib as mpl

def plot_digits(instances, images_per_row=5, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    # n_rows = ceil(len(instances) / images_per_row) 와 동일합니다:
    n_rows = (len(instances) - 1) // images_per_row + 1

    # 필요하면 그리드 끝을 채우기 위해 빈 이미지를 추가합니다:
    n_empty = n_rows * images_per_row - len(instances)
    padded_instances = np.concatenate([instances, np.zeros((n_empty, size * size))], axis=0)

    # 배열의 크기를 바꾸어 28×28 이미지를 담은 그리드로 구성합니다:
    image_grid = padded_instances.reshape((n_rows, images_per_row, size, size))

    # 축 0(이미지 그리드의 수직축)과 2(이미지의 수직축)를 합치고 축 1과 3(두 수평축)을 합칩니다. 
    # 먼저 transpose()를 사용해 결합하려는 축을 옆으로 이동한 다음 합칩니다:
    big_image = image_grid.transpose(0, 2, 1, 3).reshape(n_rows * size,
                                                         images_per_row * size)
    # 하나의 큰 이미지를 얻었으므로 출력하면 됩니다:
    plt.imshow(big_image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
    

plt.figure(figsize=(12, 9))
plt.subplot(121)
plot_digits(X_train[:25, :]) # 25개의 사진, 사진마다 모든 픽셀
plt.title("Original", fontsize=16)
plt.subplot(122)
plot_digits(X_recovered[:25, :])
plt.title("Compressed", fontsize=16)
```

```python
# 점진적 PCA
# 점진적 학습의 개념

from sklearn.decomposition import IncrementalPCA

n_batches = 100 # 100개씩 학습
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(X_train, n_batches): # 지정한 학습량만큼 학습
    inc_pca.partial_fit(X_batch)
    
X_reduced = inc_pca.transform(X_train)
```

```python
# 잘되었는지 확인

X_recovered = inc_pca.inverse_transform(X_reduced) # 차원 복원
plt.figure(figsize=(12, 9))
plt.subplot(121)
plot_digits(X_train[:25, :]) # 25개의 사진, 사진마다 모든 픽셀
plt.title("Original", fontsize=16)
plt.subplot(122)
plot_digits(X_recovered[:25, :])
plt.title("Compressed", fontsize=16)
```

```python
# 커널 PCA
# 자세한 내용은 5장 참조

# 스위스롤 데이터
X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
y = t > 6.9


# rbf커널
from sklearn.decomposition import KernelPCA
rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
X_reduced_rbf = rbf_pca.fit_transform(X)

# 선형커널
lin_pca = KernelPCA(n_components=2, kernel="linear", fit_inverse_transform=True)
X_reduced_lin = lin_pca.fit_transform(X)

# 시그모이드 커널
sig_pca = KernelPCA(n_components=2, kernel="sigmoid", gamma=0.0001, coef0=1, fit_inverse_transform=True)
X_reduced_sig = sig_pca.fit_transform(X)
```

```python
# 최고의 커널선택과 하이퍼파라미터(감마) 튜닝, 선택

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ("kpca", KernelPCA(n_components=2)),
    ("log_reg", LogisticRegression()),
])

param_grid = [{
    "kpca__gamma" : np.linspace(0.03, 0.05, 10),
    "kpca__kernel" : ["rbf", "sigmoid"]
}]

# 위애서 설정한 감마와 커널로 그리드서치 수행
grid_search = GridSearchCV(clf, param_grid, cv=3)
grid_search.fit(X, y)
```

```python
print(grid_search.best_params_)
# 최선의 감마와 커널
```

```python
# 최선의 재구성 원상 구현
# 차원 되돌리기
# 완벽히는 되돌리지 못함
# 방법 -> 데이터를 투영된 샘플로 지정, 타깃을 원본 데이터로 지정 후 훈련

rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
X_reduced = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)
```

```python
# 재구성 원상 오차 계산

from sklearn.metrics import mean_squared_error
mean_squared_error(X, X_preimage)
```
