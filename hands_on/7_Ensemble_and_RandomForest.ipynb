```python
# 앙상블 학습 -> 더 좋은 결과를 만드는 방법이라고 생각
# 여러 종류 혹은 하나의 여러 예측기나 분류기들을 통해 더 좋은 예측을 만듦
# 다양한 종류의 머신을 사용해야 예측 정확도가 더 올라감.



# 1. 투표기반 분류기
# 분류기들의 결과 중 가장 많이 예측한 클래스를 선택

# 투표기반 분류기 만들기
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier # 투표기반
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC()

# 데이터 생성
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.3, random_state=53)

# 훈련 및 테스트 생성
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) 


# 직접투표
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard'
)
voting_clf.fit(X_train, y_train)
```

```python
# 정확도 비교
# 투표기반 분류기가 더 정확도가 높아짐

from sklearn.metrics import accuracy_score
for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

```python
# 간접투표
# 각 클래스의 확률을 활용하여 예측, predict_proba가 존재해야 함.
# 확률차가 큰 투표에 비중을 더 둠
# 직접투표보다 더 좋은 정확도

log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True) # 차이1

voting_clf_soft = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting="soft" # 차이2
)
voting_clf_soft.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
for clf in (log_clf, rnd_clf, svm_clf, voting_clf_soft):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

```python
# 2. 배깅과 페이스팅
# 여러 종류의 예측기가 아닌 하나의 종류의 예측기만 활용함
# 하나의 데이터셋을 여러개로 나누어 예측기에 훈련
# 배깅 -> 데이터셋을 나눌때 중복허용
# 페이스팅 -> 데이터셋을 나눌때 중복허용 안함
# 분류시에는 통계적 최빈값, 회귀에서는 평균을 계산

# 배깅 구현
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1
)
# max_samples=100을 통해 100개의 샘플 훈련
# n_estimators = 500을 통해 결정트리 500개 생성 후 사용
# 부트스트랩 true지정으로 배깅 구현, n_jobs=-1 지정시 가용한 모든 cpu사용
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

```python
# 한개의 결정트리와 비교

tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
y_pred_tree = tree_clf.predict(X_test)

print("tree_clf", accuracy_score(y_test, y_pred_tree))
print("bag_clf", accuracy_score(y_test, y_pred))
```

```python
# 페이스팅 구현

faceting_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=False, n_jobs=-1
)
faceting_clf.fit(X_train, y_train)
y_pred_face = faceting_clf.predict(X_test)

print("tree_clf", accuracy_score(y_test, y_pred_tree))
print("face_clf", accuracy_score(y_test, y_pred_face))
```

```python
# 학습률에 따른 데이터 변화

def plot_ada_plus(X, y, ada_clf, learning_rate):
    plt.plot(X[y==0, 0], X[y==0, 1], "r.")
    plt.plot(X[y==1, 0], X[y==1, 1], "b.")

    x0, x1 = np.meshgrid(
        np.linspace(-2, 3, 400).reshape(-1, 1),
        np.linspace(-1.5, 2, 300).reshape(-1, 1)
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = ada_clf.predict(X_new)
    zz = y_predict.reshape(x0.shape)
    custom_map = ListedColormap(["#fafab0", "#9898ff", "#a0faa0"])
    plt.contourf(x0, x1, zz, cmap=custom_map)
    plt.title("learning_rate = {}".format(learning_rate), fontsize=13)
    
    # 경계선
    y_pred = ada_clf.predict(X_new).reshape(x1.shape)
    custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
    plt.contour(x0, x1, y_pred, cmap=custom_cmap2, alpha=0.8)


ada_clf1 = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=10,
    algorithm="SAMME.R", learning_rate=0.5
)
ada_clf1.fit(X_train, y_train)

ada_clf2 = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=10,
    algorithm="SAMME.R", learning_rate=1
)
ada_clf2.fit(X_train, y_train)


plt.figure(figsize = (12,8))
plt.subplot(221);
plot_ada_plus(X, y, ada_clf1, 0.5)
plt.subplot(222);
plot_ada_plus(X, y, ada_clf2, 1)
```

```python
# 2. 그레이디언트 부스팅
# 이전 예측기가 만ㄴ든 잔여오차에 대한 새로운 예측기를 학습
# 실제값에서 예측값을 뺀 결과를 다음 훈련에 타깃으로 훈련



# 데이터준비
import numpy as np
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)


# 훈련 1
from sklearn.tree import DecisionTreeRegressor
tree_reg1 = DecisionTreeRegressor(max_depth=2)
tree_reg1.fit(X, y)

X_new = np.linspace(-0.5, 0.5, 100).reshape(-1, 1)
y_pred1 = tree_reg1.predict(X_new)
```

```python
# 훈련 2
y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2)
tree_reg2.fit(X, y2)
y_pred2 = y_pred1 + tree_reg2.predict(X_new) # 이전 예측과 이번 예측을 더해줘야 함
```

```python
# 훈련 2
y3 = y2 - tree_reg1.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2)
tree_reg3.fit(X, y3)
y_pred3 = tree_reg3.predict(X_new) + y_pred1 + y_pred2
```

```python
# 최적의 트리수 찾기

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor


X_train, X_val, y_train, y_val = train_test_split(X, y)

gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=120)
gbrt.fit(X_train, y_train)

errors = [mean_squared_error(y_val, y_pred)
         for y_pred in gbrt.staged_predict(X_val)]
bst_n_estimators = np.argmin(errors) + 1

gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators=bst_n_estimators)
gbrt_best.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
X_new = np.linspace(-0.5, 0.5, 100).reshape(-1, 1)
y_pred_best = gbrt_best.predict(X_new)

plt.figure(figsize=(12, 8))
plt.plot(X, y, "b.")
plt.plot(X_new, y_pred_best, "r-")
plt.title("Best gbrt")
print("best estimators number : ", bst_n_estimators)
```

```python
# 조기 종료
# 다벗번의 반복동안 검증오차가 향상되지 않으면 훈련을 중단

gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True)

min_val_error = float("inf") # 양의 무한대
error_going_up = 0
for n_estimators in range(1, 120):
    gbrt.n_estimators = n_estimators # 추정기수를 1개부터 1씩 올림
    gbrt.fit(X_train, y_train)
    y_pred = gbrt.predict(X_val)
    val_error = mean_squared_error(y_val, y_pred)
    if val_error < min_val_error: # 추정기가 1일때 오차값이 양의 무한대보다 작은지 확인
        min_val_error = val_error
        error_going_up = 0
    else:
        error_going_up += 1
        if error_going_up == 5:
            break
            
# 확률적 그레디언트 부스팅
# 각 트리가 훈련할때 사용할 훈련샘플을 무작위로 선택
# gbrt 생성시 인자로 subsample=0.25로 설정시 25퍼 크기의 무작위 데이터로 훈련
```
